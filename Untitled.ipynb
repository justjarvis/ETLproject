```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://www.kiplinger.com/slideshow/real-estate/T006-S001-most-expensive-u-s-cities-to-live-in-2019/index.html'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_="col-md-12")
```

```python
results
```

```python
# Retrieve the parent divs for all articles
city_data = soup.find_all('div', class_='kip-data')
#     # Slice the datetime string for the da
city_data
```

```python
city_name = soup.find_all('h2')
city_name
```

```python
# Loop through results to retrieve article title, header, and timestamp of article
for i in range(20):
    print(city_name[i].text)
    print(city_data[i].text)
```

```python
# Create a DataFrame of frames using a dictionary of lists
expensive_city_df = pd.DataFrame({
    "city_name": ["Los Angeles", "Boston", "San Francisco", "D.C.", "New York City"],
    "cost_of_living": [48.2, 50.0, 96.3, 62.6, 148.5],
    "median_household_income": [55, 62, 96, 78, 80],
    "median_home_value": [550, 455, 927, 537, 915]
})
expensive_city_df
```

```python
expensive_city_df['cost_of_living'] = expensive_city_df['cost_of_living']/100
expensive_city_df['median_household_income'] = expensive_city_df['median_household_income']*1000
expensive_city_df['median_home_value'] = expensive_city_df['median_home_value']*1000
expensive_city_df
```
